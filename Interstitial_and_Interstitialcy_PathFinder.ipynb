The VoronoiInterstitialGenerator displayed below is modified to return all interstitial sites, not only the unique interstitial sites. This works very well for small symmetrical structures.

```python
class VoronoiInterstitialGenerator(DefectGenerator):
    """
    Generator for interstitials based on a simple Voronoi analysis
    """

    def __init__(self, structure, element):
        """
        Initializes an Interstitial generator using Voronoi sites
        Args:
            structure (Structure): pymatgen structure object
            element (str or Element or Species): element for the interstitial
        """
        self.structure = structure
        self.element = element

        framework = list(self.structure.symbol_set)
        get_voronoi = TopographyAnalyzer(self.structure, framework, [], check_volume=False)
        get_voronoi.cluster_nodes()
        get_voronoi.remove_collisions()
        

        # trim equivalent nodes with symmetry analysis
        struct_to_trim = self.structure.copy()
        for poss_inter in get_voronoi.vnodes:
            struct_to_trim.append(self.element, poss_inter.frac_coords, coords_are_cartesian=False)
        
        symmetry_finder = SpacegroupAnalyzer(struct_to_trim, symprec=1e-1)
        equiv_sites_list = symmetry_finder.get_symmetrized_structure().equivalent_sites
        
        ## Initializing a list for interstitial sites. [[equivalent sites in the same list],[]]
        self.uniqueEquivalent = []
        
        # do additional screening for sublattice equivalent
        # defects which may have slipped through
        pdc = PointDefectComparator()
        self.unique_defect_seq = []
        self.unique_defect_seq_poss = []
        for poss_site_list in equiv_sites_list:
            poss_site = poss_site_list[0]
            if poss_site not in self.structure:
                now_defect = Interstitial(self.structure, poss_site)
                append_defect = True
                for unique_defect in range(0,len(self.unique_defect_seq)):
                    if pdc.are_equal(now_defect, self.unique_defect_seq[unique_defect]):
                        append_defect = False
                if append_defect:
                    self.unique_defect_seq.append(now_defect)
                    self.unique_defect_seq_poss.append(poss_site)
                    
                    ## Adding the interstitial sites to the list
                    self.uniqueEquivalent.append([poss_site_list[i] for i in range(0,len(poss_site_list))])
            
        self.count_def = 0  # for counting the index of the generated defect

    def __next__(self):
        """
        Returns the next interstitial or
        raises StopIteration
        """
        if len(self.unique_defect_seq) > 0:
            inter_defect = self.unique_defect_seq.pop(0)
            inter_site = inter_defect.site
            self.count_def += 1
            site_name = "Voronoi" + str(self.count_def)
            return Interstitial(self.structure, inter_site, site_name=site_name)
        raise StopIteration
    
    ## Method for retrieving all the interstitial sites
    def get_UniqueAndEquivalent(self):
        return self.uniqueEquivalent
```

The AllInterstitials object is designed to save information of the interstitial sites in a structure

```python
class AllInterstitials:
    
    '''
    Object containing a structure with all interstitial sites of one type of atom.
    Initiaizing with the regular structure + the specie in interstitial sites.
    '''
    
    def __init__(self, structure, specie):
        self.structure = structure
        self.specie = get_el_sp(specie)
        self.makeSymStructure()
        
    def makeInterstitials(self):
        self.ints = VoronoiInterstitialGenerator(self.structure, self.specie).get_UniqueAndEquivalent()
    
    def makeIntStructure(self):
        self.makeInterstitials()
        new_Ints = []
        struct = copy.deepcopy(self.structure)
        mid_struct = copy.deepcopy(self.structure) ####
        for i in self.ints:
            eq_ints = []
            iints = [] ###
            for q in i:
                mid_struct.append('X0+', q._frac_coords)
                if len(struct.get_neighbors(q, 1)) == 0:
                    struct.append(str(q._species)[:len(str(q._species))-1], q._frac_coords)
                    eq_ints.append(q)
            if len(eq_ints) != 0:
                new_Ints.append(eq_ints)
        mid_struct.to(filename='all_iinnts.vasp',fmt='poscar')
        self.int_structure = struct
        self.ints = new_Ints
    
    def makeSymStructure(self):
        self.makeIntStructure()
        self.symm_structure = SpacegroupAnalyzer(self.int_structure, symprec=0.1).get_symmetrized_structure()
    
    
    # Return the symmetrized structure
    def getSymStructure(self):
        return self.symm_structure
    
    # Return structure with all the interstitial sites
    def getAllInterstitialsStructure(self):
        return self.int_structure
    
    # Return list with all interstitial sites. [[equivalent sites in the same list],[]]
    def getInterstitials(self):
        return self.ints
```

Object describing an interstitialcy path

```python
class InterstitialcyPath:
    
    '''
    Object describing an interstitialcy path
    Args:
        - Structure
        - Migrating Specie
        - Initial , lattice and final site or two MigrationHops
    Properties:
        - Structure
        - Migrating specie
        - Angle
        - Total length
    '''
    
    def __init__(self, structure, migrating_specie, **kwargs):
        self.structure = structure
        self.migrating_specie = get_el_sp(migrating_specie)
        if 'inti' not in kwargs:
            self.initMigHop(**kwargs)
        else:
            self.initSites(**kwargs)
        self.setAngle()
        self.setTotDist()
        
    # Initializing with sites    
    def initSites(self, **kwargs):
        self.inti = kwargs['inti']
        self.intf = kwargs['intf']
        self.lattice_site = kwargs['lattice_site']
        
    # Initializing with MigrationHop (not tested and not used (at least for now))
    def initMigHop(self,**kwargs):
        self.inti = kwargs['first_hop'].isite
        self.intf = kwargs['second_hop'].esite
        self.lattice_site = kwargs['first_hop'].esite
    
    # Finding the total distance
    def setTotDist(self):
        self.tot_dist = self.inti.distance(self.lattice_site) + self.lattice_site.distance(self.intf)
    
    # Return the total distance
    def getTotDist(self):
        return self.tot_dist
    
    # toString method
    def __str__(self):
        return 'Start: '+str(self.inti._frac_coords)+ \
               '\n Lattice site: '+str(self.lattice_site)+ \
               '\n End: '+str(self.intf._frac_coords)+ \
               '\n Total distance: '+str(self.getTotDist())+ \
               '\n Angle: '+str(self.angle)
    
    # Test if equal method. Here I guess there are room for improvement
    def __eq__(self,other):
        
        # Return True if initial, lattice and final site are the same and if the angles are the same
        if self.inti.is_periodic_image(other.inti) and \
           self.lattice_site.is_periodic_image(other.lattice_site) and \
           self.intf.is_periodic_image(other.intf) and \
           round(self.angle,4)==round(other.angle,4):
            return True
        
        # Return True if initial, lattice and final site "inversed" are the same and if the angles are the same 
        elif self.intf.is_periodic_image(other.inti) and \
             self.lattice_site.is_periodic_image(other.lattice_site) and \
             self.inti.is_periodic_image(other.intf) and \
             round(self.angle,4)==round(other.angle,4):
            return  True
        
        # Return True if total distance, angle and lattice site are the same
        elif round(self.getTotDist(),4)==round(other.getTotDist(),4) and \
             round(self.angle,4)==round(other.angle,4) and \
             self.lattice_site==other.lattice_site:
            return True
        
        else:
            return False
    
    # Didin't want to write this two times :D
    def lengthVector(self, vector):
        return np.sqrt(vector[0]**2+vector[1]**2+vector[2]**2)
    
    def vecProduct(self, vec_a, vec_b):
        return (vec_a[0]*vec_b[0])+(vec_a[1]*vec_b[1])+(vec_a[2]*vec_b[2])
    
    # Finding the angle
    def setAngle(self):
        vec_a = [self.inti.x - self.lattice_site.x, self.inti.y - self.lattice_site.y, self.inti.z - self.lattice_site.z]
        vec_b = [self.intf.x - self.lattice_site.x, self.intf.y - self.lattice_site.y, self.intf.z - self.lattice_site.z]
        angle = np.arccos((self.vecProduct(vec_a, vec_b))/(self.lengthVector(vec_a)*(self.lengthVector(vec_b))))
        if angle <= np.pi:
            self.angle = angle
        else:
            self.angle = np.pi - (angle-np.pi)
    
    # Return angle
    def getAngle(self):
        return self.angle
    
    def __lt__(self, other):
        return self.angle > other.angle
    
    def make_gif(self):
        self.makeStructures()
        interpolate = self.init_structure.interpolate(self.final_structure,nimages=7,pbc=True)
        for i,structure in enumerate(interpolate):
            structure.to(filename='./gif_intcy/gif'+str(i)+'.vasp'.format(i),fmt='poscar')
            
    # Making start and end structures (not tested, may replace create POSCAR)
    def makeStructures(self):
        index = self.structure.index(self.lattice_site)
        
        self.init_structure = copy.deepcopy(self.structure)
        self.final_structure = copy.deepcopy(self.structure)

        self.init_structure.remove_sites([index])
        self.final_structure.remove_sites([index])

        self.init_structure.append(self.inti.specie, self.inti._frac_coords)
        self.init_structure.append(self.lattice_site.specie, self.lattice_site._frac_coords)

        self.final_structure.append(self.lattice_site.specie, self.lattice_site._frac_coords)
        self.final_structure.append(self.intf.specie, self.intf._frac_coords)
```

Object descibing a the interstitialcy paths of a structure

```python
class InterstitialcyPaths:
    
    '''
    Initializes an object containing all interstitialcy paths for a structure
    Args:
        - Structure
        - Migrating Specie
        - r: max distance to neighbors 
    Properties:
        - Paths
        - Structure
        - Migrating specie
    '''
    
    def __init__(self, migrating_specie, r, **kwargs):
        self.paths = []
        self.migrating_specie = get_el_sp(migrating_specie)
        self.not_in_structure = get_el_sp('X0+')
        if 'CHGCAR' in kwargs:
            chg = chgcarInterstitials(kwargs['CHGCAR'])
            int_struct = chg.getIntsStruct()
            self.ints_structure = int_struct
            self.structure = chg.getStructure()
            self.findPaths(r)
            
        elif 'structure' in kwargs:
            self.structure = kwargs['structure']
            self.ints_structure = AllInterstitials(self.structure, self.not_in_structure).getAllInterstitialsStructure()
            self.findPaths(r)
            
        else:
            print('Error: missing either a CHGCAR file og a structure')
        
            
    
    # Finding all interstitialcy paths
    def findPaths(self, r):
        
        symm_structure = SpacegroupAnalyzer(self.structure, symprec=0.1).get_symmetrized_structure()
        
        # Going through all equivalen lattice sites in the structure
        for unique in symm_structure.equivalent_sites:
            if unique[0].specie == self.migrating_specie:
                lattice_site = unique[0]
                neighbors = self.ints_structure.get_neighbors(lattice_site, r)
                # Going through the lattice sites neighbors
                for nb1 in neighbors:
                    if str(nb1.specie) == 'X0+':
                        for nb2 in neighbors:
                            if str(nb2.specie) == 'X0+' and nb1!=nb2:
                                inti = PeriodicSite(self.migrating_specie, nb1.frac_coords, self.structure.lattice)
                                intf = PeriodicSite(self.migrating_specie, nb2.frac_coords, self.structure.lattice)
                                path = InterstitialcyPath(self.structure,self.migrating_specie, 
                                                          inti=inti, 
                                                          lattice_site=lattice_site, 
                                                          intf=intf)
                                # Add path if the path not already exist and if the angle is greater than 90
                                if path not in self.paths and (path.getAngle() > (np.pi/2-0.01)):
                                    self.paths.append(path)
    
    # Return list of paths    
    def getPaths(self):
        return self.paths
    
    # Method for creating POSCARs of all paths
    def createPOSCARs(self, N):
        if N <= len(self.paths):
            for i in range(N):
                index = self.structure.index(self.paths[i].lattice_site)
                init_structure = copy.deepcopy(self.structure)
                final_structure = copy.deepcopy(self.structure)

                init_structure.remove_sites([index])
                final_structure.remove_sites([index])

                init_structure.append(self.paths[i].inti.specie, self.paths[i].inti._frac_coords)
                init_structure.append(self.paths[i].lattice_site.specie, self.paths[i].lattice_site._frac_coords)

                final_structure.append(self.paths[i].lattice_site.specie, self.paths[i].lattice_site._frac_coords)
                final_structure.append(self.paths[i].intf.specie, self.paths[i].intf._frac_coords)

                init_structure.to(filename='Interstitialcy_init_'+str(i)+'.vasp'.format(i),fmt='poscar')
                final_structure.to(filename='Interstitialcy_final_'+str(i)+'.vasp'.format(i),fmt='poscar')

        else:
            print('Error: Not that many paths')
    
    def __len__(self):
        return len(self.paths)
    
    def __str__(self):
        return 'Number of distinct interstitialcy paths found: '+str(len(self.paths))
    
    def sort(self):
        self.paths = sorted(self.paths)
```

Object describing an interstitial path

```python
class InterstitialPath:
    
    def __init__(self, structure, isite, esite, migrating_specie):
        self.structure = structure
        self.isite = isite
        self.esite = esite
        self.migrating_specie = migrating_specie
        self.makeStructures()
        
        
    def getDist(self):
        return self.isite.distance(self.esite)
    
    
    # Making start and end structures
    def makeStructures(self):
        self.init_structure = copy.deepcopy(self.structure)
        self.final_structure = copy.deepcopy(self.structure)

        self.init_structure.append(self.migrating_specie,self.isite._frac_coords)
        self.final_structure.append(self.migrating_specie,self.esite._frac_coords)
    
    def getRadius(self):
        neighbors = {}
        radius = []
        closest_neighbors = []
        interpolate = self.init_structure.interpolate(self.final_structure,nimages=20,pbc=True)
        for structure in interpolate:
            shortest = 10
            closest = 0
            nbs = structure.get_neighbors(structure[len(structure)-1], 3)
            for nb in nbs:
                dist = structure[len(structure)-1].distance(nb)
                if dist < shortest:
                    shortest = dist
                    closest = nb
                if nb not in neighbors:
                    neighbors[nb] = dist
                elif nb in neighbors and neighbors[nb] > dist:
                    neighbors[nb] = dist

            radius.append(shortest)
            closest_neighbors.append(closest)
        return min(radius)
    
    def __str__(self):
        return (str(self.migrating_specie)+'   '+str(self.isite._frac_coords)+'   '+str(self.esite._frac_coords)+'   ')
    
    def __lt__(self, other):
        return self.getRadius() > other.getRadius()
    
    def make_gif(self):
        interpolate = self.init_structure.interpolate(self.final_structure,nimages=7,pbc=True)
        for i,structure in enumerate(interpolate):
            structure.to(filename='./gif/gif'+str(i)+'.vasp'.format(i),fmt='poscar')
        
```

Object descibing a the interstitial paths of a structure

```python
class InterstitialPaths:
    
    '''
    Initializes an object containing all interstitial paths for a structure
    Args:
        - Structure
        - Migrating Specie
        - r: max length of path 
    Properties:
        - Paths
        - Structure
        - Migrating specie
    '''
    
    def __init__(self, migrating_specie, r, **kwargs):
        self.migrating_specie = get_el_sp(migrating_specie)
        self.paths = []
        if 'CHGCAR' in kwargs:
            chg = chgcarInterstitials(kwargs['CHGCAR'])
            self.ints_structure = chg.getIntsStruct()
            self.structure = chg.getStructure()
            self.findPathsCHGCAR(r)

        elif 'structure' in kwargs:
            self.structure = kwargs['structure']
            self.findPaths(r)
            
        else:
            print('Error: missing either a CHGCAR file og a structure')
    
    
    def findPathsCHGCAR(self,r):
        paths = pathfinder.DistinctPathFinder(self.ints_structure,'X0+')
        paths = paths.get_paths()
        filtered_paths = []
        for path in paths:
            if path.isite.distance(path.esite) < r and path.isite.distance(path.esite) > 1.5:
                new_path = InterstitialPath(self.structure, path.isite, path.esite, self.migrating_specie)
                filtered_paths.append(new_path)
        self.paths = filtered_paths
    
    # Finding interstitial paths
    def findPaths(self, r):
        
        AllInts = AllInterstitials(self.structure, self.migrating_specie)
        
        all_ints = AllInts.getAllInterstitialsStructure()
        symm_structure = AllInts.getSymStructure() # Symmetrized version of the structure with all interstitials
        ints = AllInts.getInterstitials() # List of all interstitial sites
        
        paths = []
        
        # Finding the shortest path between all sets of equivalent sites
        for list_1_of_eqs in ints:
            for list_2_of_eqs in ints:
                something_to_append = False
                shortest = 100 # Random big number
                temp_start = 0 # Just need to initialize with something
                temp_end = 0 # Just need to initialize with something
                for eq1 in list_1_of_eqs:
                    for eq2 in list_2_of_eqs:
                        dist = eq1.distance(eq2)
                        if dist != 0 and dist < shortest:
                            something_to_append = True
                            shortest = dist
                            temp_start = eq1
                            temp_end = eq2
                if something_to_append:
                    path = InterstitialPath(self.structure, temp_start, temp_end, self.migrating_specie)
                    #path = MigrationHop(temp_start, temp_end, symm_structure)
                    if shortest < r and shortest > 0.5: # Maybe not nessesary to include the path if it is very short
                        paths.append(path)
        
        # Check for duplicate paths
        for path in paths:
            test = True
            for i in self.paths:
                if (path.isite.is_periodic_image(i.isite) and path.esite.is_periodic_image(i.esite)) or \
                   (path.isite.is_periodic_image(i.esite) and path.esite.is_periodic_image(i.isite)):
                    test = False
            if test:
                self.paths.append(path)
        
    def getPaths(self):
        return self.paths
    
    
    # Method to create POSCARs for all paths in this object
    def createPOSCARs(self, N):
        
        if N <= len(self.paths):
            index = 0
            for i in range(N):

                self.paths[i].init_structure.to(filename='Interstitial_init_'+str(index)+'.vasp'.format(index),fmt='poscar')
                self.paths[i].final_structure.to(filename='Interstitial_final_'+str(index)+'.vasp'.format(index),fmt='poscar')

                index += 1
        else:
            print('The number of paths chosen are too high')
    
    # toString method
    def __str__(self):
        return 'Number of distinct interstitial paths found: '+str(len(self.paths))
    
    def findPathRadius(self):
        self.radius_of_paths = []
        for path in self.paths:
            self.radius_of_paths.append(path.getRadius())
    
    def sort(self):
        self.paths = sorted(self.paths)
                
```

Object for interstitial sites from a CHGCAR file

```python
class chgcarInterstitials:
    
    def __init__(self, chg):
        chg = ChargeDensityAnalyzer(chg)
        chg.get_local_extrema()
        chg.remove_collisions(1.2)
        chg.cluster_nodes(1.2)
        int_struct = chg.get_structure_with_nodes()
        int_struct.to(filename='entesttil.vasp',fmt='poscar')
        
        self.structure = copy.deepcopy(int_struct)
        
        self.l_ints = []
        counter = 0
        for i in int_struct:
            if str(i.specie) == 'X0+':
                self.l_ints.append(i)
                self.structure.__delitem__(counter)
            else:
                counter += 1
        self.updateInts()
        
    def updateInts(self):
        self.new_IntStruct = copy.deepcopy(self.structure)
        
        for i in self.l_ints:
            midstruct = copy.deepcopy(self.new_IntStruct)
            midstruct.append(i.specie, i._frac_coords)
            nbs = midstruct.get_neighbors(i,1)
            if len(nbs) == 0:
                self.new_IntStruct.append(i.specie, i._frac_coords)
    
    def getIntsStruct(self):
        return self.new_IntStruct
    
    def getStructure(self):
        return self.structure
            
```
